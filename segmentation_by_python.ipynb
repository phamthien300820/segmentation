```python
import numpy as np, pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('segmentation_v2.csv')
print (df)
```

```python
df.describe()
```

```python
age_group = pd.cut(df.age, bins = [-1,18,26,32,40,100], labels= ['18-','18-25','26-31','32-39','40+'])
df.insert(4, 'age_group', age_group)
df.sample(10)
```

```python
year_exp_group = pd.cut(df.year_experience, bins = [-1,1,5,10,100], labels= ['0','1-4','5-9','10+'])
df.insert(6, 'experience_group', year_exp_group)
current_salary_group = pd.cut(df.salary_current, bins = [-1,1,1000,2000,3000,100000], labels= ['0','1-999','1000-1999','2000-2999','3000+'])
df.insert(13, 'current_salary_group', current_salary_group)
expected_salary_group = pd.cut(df.salary_expected, bins = [-1,1,1000,2000,3000,100000], labels= ['0','1-999','1000-1999','2000-2999','3000+'])
df.insert(15, 'expected_salary_group', expected_salary_group)
df.sample(10)
```

```python
df_train = df_train.replace(np.nan,0)
df_train
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(4)

train = df_train.iloc[:,:]
train
```

```python
identified_clusters = kmeans.fit_predict(train)
identified_clusters
```
